```python
import pandas as pd
import numpy as np
import scipy.stats
```

```python
df = pd.read_csv('subject_dependent_model_metrics.csv')
```

```python
df
```

```python
def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h
```

```python
print(mean_confidence_interval(df['overall_accuracy']))
print(mean_confidence_interval(df['valence_accuracy']))
print(mean_confidence_interval(df['arousal_accuracy']))
```
